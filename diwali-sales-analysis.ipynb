## Q9. Plot a bar chart for Gender and it's count 

```python
plt.figure(figsize=(5,3))
sns.countplot(x="Gender",data=df, )
plt.title("Distribution of Gender")
plt.tight_layout()
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

## Q10. Plot a bar chart for gender vs total amount 

```python
plt.figure(figsize=(5,3))
sns.barplot(x="Gender",data=df,y="Amount" )
plt.title("Distribution of Gender vs Amount")
plt.tight_layout()
plt.xlabel("Gender")
plt.ylabel("Amount")
plt.show()
```

## Q11. Plot the countplot for Age Group with hue and Gender

```python
plt.figure(figsize=(5,3))
sns.countplot(x="Age Group",data=df,hue ="Gender" )
plt.title("Distribution of Gender vs Age group")
plt.tight_layout()
plt.xlabel("Age Group")
plt.ylabel("Gender")
plt.show()
```

## Q12. Draw the plot for total number of orders from top 10 states 

```python
top_states = df['State'].value_counts().head(10)
plt.figure(figsize=(15,5))
sns.barplot(x=top_states.index,y=top_states.values, palette='viridis')
plt.title("Top 10 sates of max orders")
plt.tight_layout()
#plt.xticks(rotation=45)
plt.xlabel("State")
plt.ylabel("No. of Orders")
plt.tight_layout()
plt.show()
```

## Q13.Draw the plot to plot total amount/sales from top 10 states

```python
sales_by_state = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=sales_by_state.index, y=sales_by_state.values, palette='magma')

plt.title('Top 10 States by Total Sales')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Q14. Draw the plot to show the marital status 

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(data=df, x='Marital_Status',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
#plt.title("Marital Status")
plt.tight_layout()
plt.show()
```

## Q15. Draw the plot to show the top 10 occuoations

```python
occ = df['Occupation'].value_counts().head(10)
plt.figure(figsize=(8,5))
ax = sns.barplot(x=occ.index,y=occ.values, palette='magma')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Top 10 occupation")
plt.tight_layout()
plt.xticks(rotation=45)
plt.xlabel("Occupations")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

## Q16. Draw the countplot to print the top 10 categories 

```python
category = df['Product_Category'].value_counts().head(10)
plt.figure(figsize=(8,6))
ax = sns.countplot(x=category.index,palette='viridis')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Top 10 Product category")
plt.tight_layout()
plt.xticks(rotation=75)
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

## Q17. What is the prodct ID of the top 10 most sold products.

```python
s = df['Product_ID'].value_counts().head(10).reset_index()
s.columns=['Product_ID','Count']
plt.figure(figsize=(10,4))
ac = sns.barplot(data = s,x='Product_ID',y='Count',palette='viridis')
for i in ac.containers:
    ac.bar_label(i)
plt.xticks(rotation=75)
plt.show()
```
